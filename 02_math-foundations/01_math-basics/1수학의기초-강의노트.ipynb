## p.52

$f(x) = x^3 + 1. g(x) = \sqrt(x+2)$로 정의된 함수 $f,g$에 대해서 $(g \circ f)(2)$, $(f \circ g)(2)$, $(g \circ g)(2)$를 각각 구하기

```python
def f(x):
    return x**3+1
def g(x):
    return np.sqrt(x+2)
print("(g o f)(2)=", g(f(2)))
print("(f o g)(2)=", f(g(2)))
print("(g o g)(2)=", g(g(2)))
```

## p.57

$\sigma(x) = \frac{1}{1+exp(-x)}$

```python
### 로지스틱 함수 ###
def logistic(x):
    return 1/(1+np.exp(-x))

xx = np.linspace(-10,10,100)
plt.plot(xx, logistic(xx))
plt.title('로지스틱함수')
plt.xlabel('$x$') # x축
plt.ylabel('$\sigma(x)$') # y축
plt.show()
```

## p.60

$\zeta(x) = \log(1+\exp(x))$

```python
### 소프트플러스 함수 ###
def softplus(x):
    return np.log(1+np.exp(x))

xx = np.linspace(-10,10,100)
plt.plot(xx, softplus(xx))
plt.title('소프트플러스함수')
plt.xlabel('$x$') # x축
plt.ylabel('Softplus($x$)') # y축
plt.show()
```

## p.61

```python
### 다변수함수 ###
def f(x,y):
    return 2 * x**2 + 6*x*y + 7 * y**2 - 26*x - 54*y + 107

xx = np.linspace(-3,7,100)
yy = np.linspace(-3,7,100)
X, Y = np.meshgrid(xx,yy)
Z = f(X,Y)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X,Y,Z, linewidth=0.1)
ax.view_init(40,-110)
plt.xlabel('x')
plt.ylabel('y')
plt.title('서피스플롯(Surface plot)')
plt.show()
```

```python
CS = plt.contour(X,Y,Z, levels = np.logspace(0,3,10))
plt.clabel(CS, fmt = '%d')
plt.title('컨투어플롯')
plt.show()
```
