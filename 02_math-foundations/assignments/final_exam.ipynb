#### (a) 세토사 종 (첫 50개의 관측치)의 붓꽃 중 각 변수 (꽃받침 길이, 꽃받침 폭, 꽃잎 길이, 꽃잎 폭)의 최댓값을 구하라.

#### (b) 첫 3개의 관측치를 벡터 a, b, c 로 각각 표현하고, 각 벡터의 유클리디안 길이를 구하라.

```python
# 세 개의 벡터로 표현
a = np.array(df.iloc[0])
b = np.array(df.iloc[1])
c = np.array(df.iloc[2])
print( f" a = {a} \n b = {b} \n c = {c}")

# 각 벡터의 유클리디안 길이 계산
print(f" a의 유클리디안 길이: {np.linalg.norm(a)}")
print(f" b의 유클리디안 길이: {np.linalg.norm(b)}")
print(f" c의 유클리디안 길이: {np.linalg.norm(c)}")
```

(c) 위의 (b)의 벡터들 사이의 코사인 거리를 구하라. 어느 관측치 쌍이 가장 가까운가?

```python
def cosine_distance(vec1, vec2):  
    dot_product = np.dot(vec1, vec2)  
    norm_vec1 = np.linalg.norm(vec1)  
    norm_vec2 = np.linalg.norm(vec2)  
    if norm_vec1 == 0 or norm_vec2 == 0:  
        return np.nan  
    return 1 - (dot_product / (norm_vec1 * norm_vec2))

cosine_ab = cosine_distance(a, b)
cosine_ac = cosine_distance(a, c)   
cosine_bc = cosine_distance(b, c)   

print(f" a와 b의 코사인 거리: {cosine_ab}")
print(f" a와 c의 코사인 거리: {cosine_ac}")
print(f" b와 c의 코사인 거리: {cosine_bc}") 

pairs = {'(a, b)' : cosine_ab, '(a, c)' : cosine_ac, '(b, c)' : cosine_bc}
closest = min(pairs, key=pairs.get)
print(f" 가장 가까운 관측치 쌍: {closest}")

# 참고 a,c 의 코사인 거리 scipy 라이브러리를 이용한 계산
#from scipy.spatial import distance
#print( distance.cosine(a, c) )
```
