### p.19 함수의 극한 연습문제

[문제 1] 다음 명제의 참/거짓 판별 : $lim_{x \to \infty} \{f(x) + g(x)\} = lim_{x \to \infty} f(x) + lim_{x \to \infty} g(x)$

정답 : 거짓

이유 : 해당 명제는 
$lim_{x \to \infty} f(x) $ 와
$lim_{x \to \infty} g(x) $ 가 각각 수렴할 때만 성립함

[문제 2] $lim_{x \to 0} \frac{|x|}{x}$ 을 구하라.

정답 : 극한값이 존재하지 않음

이유 : 우극한($x \to 0^+$)은 1이고, 좌극한($x \to 0^-$)은 -1이므로 좌우극한이 일치하지 않음



[문제 3] $lim_{x \to -5} \frac{x^2 - 25}{x+5}$ 을 구하라.

풀이 :  $\frac{x^2 - 25}{x+5}$ = $\frac{(x-5)(x+5)}{x+5} = x-5$  이고, $x = -5$를 대입해 주면 $-5 -5 = -10$

정답 : -10 

### p.26 함수의 극한 연습문제

[문제 1] 함수 $f(x)$가 구간 (0,1)에서 연속이면, 최댓값과 최솟값을 갖는가?

정답 : 거짓 

이유 : 최대최소정리는 닫힌 구간에서만 성립되며, 
열린 구간에서는 양 끝 점으로 수렴할 뿐 함수가 최댓값과 최솟값을 갖는다고 보장할 수 없음

[문제 2] $f(x) = \frac{x+4}{x-1}$은 $x=1$에서 연속인가?

정답 : 불연속

이유 : $x = 1$ 에서 함수값 $f(1)$ 이 정의되지 않으므로 불연속이다. (분모 = 0)

[문제 3] $x^3 = x - 3$이 (-2, 1)에서 적어도 하나의 실근을 가짐을 증명하라.

증명 : $f(x) = x^3 - x + 3$ 이라고 할 때, $f(-2) = -8 + 2 + 3 = -3$ 이고, $f(1) = 1-1+3 = 3$이다.

$f(-2) < 0 < f(1)$이며 $f(x)$는 연속함수이므로, 중간값 정리에 의해 $f(c) = 0$인 $c \in (-2,1)$가 적어도 하나 존재함

##### [문제 4] 일반항 유도
1. 주어진 식의 양 변에 역수를 취하면 : $\frac{1}{a_{n+1}} = \frac{2a_{n}+1}{a_{n}} = 2 + \frac{1}{a_{n}}$
2. 따라서 $b_{n} = \frac{1}{a_{n}}$ 라고 하면, $b_{n+1} = b_{n} + 2 $ 가 되어 등차수열이 됨
3. $b_{1} = \frac{1}{a_{1}} = 1$ 이므로, $b_{n} = 1 + 2(n-1) = 2n - 1 $ 이 됨
4. 따라서 $a_{n} = \frac{1}{2n - 1}$ 이다.

```python
def get_a_n(n) :
    return 1 / (2 * n - 1)

a_100 = get_a_n(100)
print(f"[문제 4] a_100 = {a_100} (분수 : 1/199)")
```

### p.33 도함수 연습문제

##### [문제1] $f(x) = \frac{e^x}{x}$ ($x \neq 0$)


참고. 몫의 미분법 적용 : $\displaystyle \left( \frac{f(x)}{g(x)} \right)' = \frac{f'(x)g(x) - f(x)g'(x)}{\{g(x)\}^2}$

$u = e^x, v = x$라고 하면 $u' = e^x, v' = 1$

$$\begin{aligned} f'(x) &= \frac{(e^x)' \cdot x - e^x \cdot (x)'}{x^2} \\ &= \frac{e^x \cdot x - e^x \cdot 1}{x^2} \\ &= \frac{xe^x - e^x}{x^2} \\ &= \frac{x-1}{x^2}e^x \end{aligned}$$

##### [문제2] $g(x) = 3 \log_2 x - x \log x$ 

로그의 미분법 $\displaystyle (\log_a x)' = \frac{1}{x \log a}$ 과 

곱의 미분법 $\displaystyle (fg)' = f'g + fg'$  을 적용

$$\begin{aligned} g'(x) &= 3 \cdot \left( \log_2 x \right)' - \left( x \log x \right)' \\ &= 3 \cdot \frac{1}{x \ln 2} - \left( (x)' \log x + x (\log x)' \right) \\ &= \frac{3}{x \ln 2} - \left( 1 \cdot \log x + x \cdot \frac{1}{x} \right) \\ &= \frac{3}{x \ln 2} - \log x - 1 \end{aligned}$$



##### [문제3] $h(x) = 2 \cos(x) - \tan(x)$
$(\cos x)' = -\sin x$ 

$(\tan x)' = \sec^2 x = \frac{1}{\cos^2 x}$

$$\begin{aligned} h'(x) &= 2(-\sin x) - \sec^2 x \\ &= -2\sin x - \sec^2 x \end{aligned}$$

##### [문제4] $z(x) = e^x \log x$

$$\begin{aligned} z'(x) &= (e^x)' \log x + e^x (\log x)' \\ &= e^x \log x + e^x \cdot \frac{1}{x} \\ &= e^x \left( \log x + \frac{1}{x} \right) \\ &= \frac{x \log x + 1}{x} e^x \end{aligned}$$

```python
import sympy as sym
x = sym.Symbol('x')

# 각 함수 정의 및 미분 실행
f = sym.exp(x) / x
g = 3 * sym.log(x, 2) - x * sym.log(x)
h = 2 * sym.cos(x) - sym.tan(x)
z = sym.exp(x) * sym.log(x)

results = {
    "f'(x)": sym.simplify(sym.diff(f, x)),
    "g'(x)": sym.simplify(sym.diff(g, x)),
    "h'(x)": sym.simplify(sym.diff(h, x)),
    "z'(x)": sym.simplify(sym.diff(z, x))
}

for label, res in results.items():
    print(f"{label} = {res}")
```
