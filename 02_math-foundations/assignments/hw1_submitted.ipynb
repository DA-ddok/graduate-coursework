## 1. 집합의 합과 곱 증명
**문제:** 두 집합 $X_1, X_2$가 각각 3개의 원소를 가질 때, 다음 식의 좌변과 우변이 같음을 증명
$$\prod_{i=1}^{2} \sum_{X_i} x_i = \sum_{X_1 \times X_2} \prod_{i=1}^{2} x_i$$

**증명:**
집합 $X_1 = \{a_1, a_2, a_3\}$, $X_2 = \{b_1, b_2, b_3\}$라고 가정하였음

1. **좌변(LHS)**
   $$\prod_{i=1}^{2} \sum_{X_i} x_i = \left( \sum_{x \in X_1} x \right) \cdot \left( \sum_{y \in X_2} y \right)$$
   $$= (a_1 + a_2 + a_3)(b_1 + b_2 + b_3)$$
   $$= a_1b_1 + a_1b_2 + a_1b_3 + a_2b_1 + a_2b_2 + a_2b_3 + a_3b_1 + a_3b_2 + a_3b_3$$

2. **우변(RHS)**

   $X_1 \times X_2$는  카테시안 곱으로 형성되어, ${ (a_1, b_1), (a_1, b_2), \dots, (a_3, b_3) }$인 9개의 원소를 가진 집합이므로

   $$\sum_{X_1 \times X_2} \prod_{i=1}^{2} x_i = \sum_{(x, y) \in X_1 \times X_2} (x \cdot y)$$
   $$= a_1b_1 + a_1b_2 + a_1b_3 + a_2b_1 + a_2b_2 + a_2b_3 + a_3b_1 + a_3b_2 + a_3b_3$$

따라서 주어진 등식은 성립함

## 2. 넘파이를 사용해 붖꽃 데이터에 대한 행렬변수 $X$를 만들고 전치행렬 $X^t$ 구하기

```python
import numpy as np
from sklearn.datasets import load_iris

# 1. 붓꽃 데이터 로드
iris = load_iris()

# 2. 넘파이 행렬 X 생성 (150개 샘플, 4개 특성)
X = np.array(iris.data)
#print(X)

# 3. 전치행렬 XT 구하기
XT = X.T

# 결과 출력
print(f"1. 원본 행렬 X의 크기: {X.shape}")
print(f"2. 전치 행렬 XT의 크기: {XT.shape}")
```

## 3. 벡터 및 행렬 만들고 넘파이로 표현하기

```python
# 1. 영벡터 (Zero Vector)
zero_v = np.zeros(3)

# 2. 일벡터 (Ones Vector)
one_v = np.ones((3, 1))

# 3. 정방행렬 (Square Matrix)
sq_mat = np.array([[1, 2], [3, 4]])

# 4. 대각행렬 (Diagonal Matrix)
diag_mat = np.diag([10, 20, 30])

# 5. 항등행렬 (Identity Matrix)
identity_mat = np.identity(3)

# 6. 대칭행렬 (Symmetric Matrix, X == X.T)
sym_mat = np.array([[1, 2, 3], [2, 5, 6], [3, 6, 9]])
sym_mat_T = sym_mat.T

print(f"영벡터: {zero_v}")
print(f"일벡터:\n{one_v}")
print(f"정방행렬:\n{sq_mat}")
print(f"대각행렬:\n{diag_mat}")
print(f"항등행렬:\n{identity_mat}")
print(f"대칭행렬:\n{sym_mat}")
print(f"대칭행렬의 전치:\n{sym_mat_T}")
```

## 4. 지수 법칙 연산
**문제:** 실수 $x$에 대하여 $4^x = 7$일 때, $\frac{2^x - 2^{-x}}{2^x + 2^{-x}}$를 구하라.

**풀이 과정:**
1. 주어진 조건 $4^x = 7$을 이용하기 위해 식을 변형 > 분수식의 분모와 분자에 $2^x$를 곱합니다.
   $$\frac{2^x - 2^{-x}}{2^x + 2^{-x}} \times \frac{2^x}{2^x} = \frac{(2^x)^2 - 2^0}{(2^x)^2 + 2^0}$$
2. $(2^x)^2 = 4^x$ 이고 $2^0 = 1$이므로 식을 대입합니다.
   $$\frac{4^x - 1}{4^x + 1} = \frac{7 - 1}{7 + 1} = \frac{6}{8} = \frac{3}{4}$$

**답:** $0.75$


```python
# 방법 2: 넘파이를 이용한 수치 대입 (검증용)
import numpy as np

# 2^x = sqrt(7)임을 활용
two_pow_x = np.sqrt(7)
val_numerical = (two_pow_x - 1/two_pow_x) / (two_pow_x + 1/two_pow_x)

print(f"수치 대입 결과: {val_numerical}")
```

## 5. 테일러 정리

```python
import numpy as np
import matplotlib.pyplot as plt

#1. 데이터 준비 : -pi부터 pi까지 100개의 점 생성
x = np.linspace(-np.pi, np.pi, 100)
y_sin = np.sin(x)

#2. 주어진 수식 대입 (테일러 다항식)
t1 = x
t2 = x - (x**3)/6
t3 = x - (x**3)/6 + (x**5)/120
t4 = x - (x**3)/6 + (x**5)/120 - (x**7)/5040

#3. 그래프 시각화 
plt.figure(figsize = (10, 6))

# sin(x) 그래프
plt.plot(x, y_sin, label='sin(x)', color='blue')   

# 테일러 다항식 그래프 (점선)
plt.plot(x, t1, '--', label='T1 (x) = x')
plt.plot(x, t2, '--', label='T2 (x) = x - x^3/6')
plt.plot(x, t3, '--', label='T3 (x) = x - x^3/6 + x^5/120')
plt.plot(x, t4, '--', label='T4 (x) = x - x^3/6 + x^5/120 - x^7/5040')    

# 그래프 설정
plt.title('Approximation of sin(x) using Taylor Polynomials')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1.5, 1.5)
plt.legend()
plt.grid(True)
plt.show()
```
