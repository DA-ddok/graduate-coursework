## p.6 수열의 합

$a_n = n$일 때 $S_n = \sum_{i=1}^n a_i$ 구하는 식을 구현하고 $S_{10}$을 계산해보자.

```python
#파이썬 코드
N = 10
a = 1

sum1 = 0
for n in range(1,N+1):
    sum1 = sum1 + n

print("S10 =", sum1)
```

$a_n = 2 \times 3^{n-1}$일 때, $S_{10}$을 계산해보자.

```python
# 파이썬 코드
N = 10
a = 2

sum1 = 0
for n in range(1,N+1):
    sum1 = sum1 + a*3**(n-1)

print("S10 =", sum1)

2*(3**(10)-1)/(3-1)
```

## p.7 수열의 합 연습문제

$a_n = 5+(n-1)3$에 대하여 $S_n = \sum_{i=1}^n a_i$구하는 식을 구현하고 $S_{10}, S_{20}$을 계산해보자.


```python
#파이썬 코드
N = 10
a = 5
d = 3

sum1 = 0
for n in range(1,N+1):
    sum1 = sum1+(a+(n-1)*d)

print("S10 =", sum1)

### S_20 구하려면 N=20으로 설정 ###
```

## p.13 등비급수의 수렴과 발산

```python
### 등비급수 수렴 시뮬레이션 예1)
sum1 = 0
n = 1
diff = 1
while(diff > 1e-6):
    sum0 = sum1 + (1/2)**(n-1)
    diff = abs(sum0-sum1)
    n = n + 1
    sum1 = sum0

sum0
```

## p.68 퀴버플롯

```python
### quiver plot
def f(x,y):
    return 2*x**2+6*x*y+7*y**2-26*x-54*y+107

xx = np.linspace(1,16,100)
yy = np.linspace(-3,6,90)
X,Y = np.meshgrid(xx,yy)
Z=f(X,Y)
```

```python
def gx(x,y):
    return 4*x+6*y-26

def gy(x,y):
    return 6*x+14*y-54
```

```python
xx2 = np.linspace(1,16,15)
yy2 = np.linspace(-3,6,9)
X2, Y2 = np.meshgrid(xx2,yy2)
GX = gx(X2,Y2)
GY = gy(X2,Y2)
plt.figure(figsize = (10,5))
plt.contour(X,Y,Z, levels = np.logspace(0,3,10))
plt.quiver(X2,Y2,GX,GY, color='blue', scale=400, minshaft=2)

plt.xlabel('x')
plt.ylabel('y')
plt.title('퀴버 플롯(quiver plot)')
plt.show()
```
