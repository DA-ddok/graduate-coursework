## 문제 1: Temperature Classifier (if-statement, 6 points)

#### Objective: Write a program that classifies temperatures into different categories.

#### Requirements:
1. Ask the user to input a temperature in Celsius
2. Classify it according to:
   - `Freezing` if below 0
   - `Cold` if between 0-15 (inclusive)
   - `Mild` if between 16-25 (inclusive)
   - `Hot` if above 25
3. Print the classification

#### Example Output:
```python
Enter temperature in Celsius: 12
Classification: Cold
```

````python
#1. Ask the user to input a temperature in Celsius
temp_input = input("Enter temperature in Celsius: ")
temp = float(temp_input)

#2. Classify
if temp < 0:
    classification = 'Freezing'
elif 0 <= temp <= 15 :
    classification = 'Cold'
elif 16 <= temp <= 25 :
    classification = 'Mild'
else :
    classification = 'Hot'

#3. Print the classification
print("Classification: {}".format(classification))
````

## 문제 2: Leap Year Checker (if-statement, 6 points)


#### Write a program that:

1. Takes a year as input.
2. Checks if it is a leap year (divisible by 4, not by 100, unless by 400).
3. Prints `Leap year` or `Not a leap year`.

#### Example Output:
```python
Enter a year: 2024  
Leap year!  
```

````python
#1. Takes a year as input
year = int(input("Enter a year: "))

#2~2
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap year!")
else:
    print("Not a leap year")
````

## 문제 3: Number Guessing Game (while-loop, 7 points)

#### Write a program that:
1. Generates a random `target` number between 1-100
2. Uses a while loop to:
   - Ask for user's guess
   - Provide `Too high` or `Too low` feedback
   - Count `attempts`
3. When guessed correctly, shows `attempts` count

#### Example Output:

```python
Guess (1-100): 50
Too high!
Guess again: 25
Correct! Attempts: 2
```

````python
import random

#1. Generates a random target number between 1-100
target = random.randint(1,100)
attempts = 0

#2. while 루프
while True:

    # 첫 시도와 이후 입력 구분
    if attempts == 0:
        my_guess = int(input("Guess (1-100): "))
    else:
        my_guess = int(input("Guess again: "))

    # 시도 횟수 증가
    attempts += 1

#3. When guessed correctly, shows attempts count

    if my_guess == target :
        print(f"Correct! Attempts: {attempts}")
        break
    elif my_guess > target :
        print("Too high!")
    else:
        print("Too low!")
````

## 문제 4: Vowel Counter (for-loop, 7 points)

#### Write a program that:
1. Takes a string input
2. Counts vowels (a,e,i,o,u, case-insensitive)
3. Prints the vowel count

#### Example
```python
Enter text: "Hello World"
Vowel count: 3
```

````python

#1. Takes a string input
my_string =  input("Enter text: ")

#2. Counts vowels (a,e,i,o,u, case-insensitive)
text = my_string.lower()
vowels = "aeiou"
count = 0

for char in text:
    if char in vowels:
        count +=1

#3. Prints the vowel count
print(f"Vowel count: {count}")
````

## 문제 5: Multiplication Table (for-loop, 7 points)

#### Write a program that:
1. Takes an integer input `n`
2. Uses a for loop to print its multiplication table (1-10)
3. Formats output as `n x i = result`

#### Example Output:
```python
Enter number: 5
5 x 1 = 5
5 x 2 = 10
5 x 3 = 15
...
5 x 10 = 50
```

````python
#1. Takes an integer input n
n = int(input("Enter number: "))
#print("Enter number: {}".format(n))

#2. Uses a for loop to print its multiplication table (1-10)
for i in range(1,11):
    result = n * i
    print(f"{n} x {i} = {result}")

````

## 문제 6: Student Grade Analyzer (Pandas, 7 points)


#### Write a program that:
1. Creates a DataFrame with:
   - Name (input)
   - Score (input 0-100)
   - Grade (calculated)
2. Calculates grades (A: 90-100, B: 80-89, C: 70-79, D: 60-69, F: \<60)
3. Prints DataFrame and average score

#### Example Output:
```python
      Name  Score Grade
0    Alice     92     A
1      Bob     78     C
2  Charlie     85     B
Average score: 85.0
```

````python
import pandas as pd

#1. Creates a DataFrame
data = {
    'Name' : ['Alice','Bob','Charlie'],
    'Score' : [92,78,85]
}
df = pd.DataFrame(data)

#2. 등급 계산
def get_grade(score):
    if score >= 90:   return 'A'
    elif score >= 80: return 'B'
    elif score >= 70: return 'C'
    elif score >= 60: return 'D'
    else: return 'F'

df['Grade'] = df['Score'].apply(get_grade)

#3. 결과출력
print(df)
print(f"Average score: {df['Score'].mean()}")
````
