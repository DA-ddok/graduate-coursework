# 2. 레이아웃 - 앱 꾸미기

```python
# Initialize the app
app = Dash(__name__)

app.layout = html.Div(children = [
    html.H1(children = 'Hello World'),
    html.Div(children = 'This is my first appllication.')
])

# Run the app
if __name__ == '__main__':
    app.run(debug = True)
```

```python
# Initialize the app
app = Dash(__name__)

app.layout = html.Div(children=[
    html.H1(children = 'Hello World',
            style = {'textAlign': 'center','color': 'blue'}),

    html.Div(children = 'This is my first application.',
             style = {'textAlign': 'left', 'color': 'red'})
])

# Run the app
if __name__ == '__main__':
    app.run(debug = True)
```

```python
# htlm 대신 마크다운 사용 가능
app = Dash(__name__)

markdown_text = '''
### Dash를 이용한 어플만들기

마크다운을 활용해 봅시다! \n
this is your first introduction to Markdown!
'''

app.layout = html.Div([
    dcc.Markdown(children = markdown_text)
])

if __name__ == '__main__':
    app.run(debug = True)
```

# 3. 데이터와 연결하기

```python
!pip install -U finance-datareader

import FinanceDataReader as fdr
from datetime import datetime, timedelta

start_day = datetime(2024, 1, 1)

samsung_stock = fdr.DataReader('005930', start_day)
samsung_stock = samsung_stock.rename_axis('date').reset_index()
samsung_stock.head()
```

```python
# dash.dash_table.DataTable
# Initialize the app
app = Dash()

# App layout
app.layout = [
    html.Div(children = '2024년 삼성전자 주가'),
    dash_table.DataTable(data = samsung_stock.to_dict('records'),
                         page_size = 10)
]

# Run the app
if __name__ == '__main__':
    app.run(debug = True)
```

# 4. Plotly 객체 연결하기

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Candlestick(
    x = samsung_stock.index,
    open = samsung_stock['Open'],
    close = samsung_stock['Close'],
    high = samsung_stock['High'],
    low = samsung_stock['Low'],
    increasing = dict(line = dict(color = 'red')),   # 상승 색
    decreasing = dict(line = dict(color = 'blue')))) # 하락 색

fig.update_layout(xaxis = dict(rangeslider = dict(visible = False)))

fig.show()
```

```python
# dash.dcc.Graph
app = Dash()

app.layout = html.Div(
    [
        html.H1("2024년 삼성전자 주가 캔들차트",
               style = {'textAlign': 'center'}),
        dcc.Graph(figure = fig),
    ]
)

if __name__ == "__main__":
    app.run_server(debug = True)
```
