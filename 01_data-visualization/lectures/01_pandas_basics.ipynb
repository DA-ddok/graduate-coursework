## Pandas 핵심기능

### 1. 재색인 (Reindex)

- 새로운 색인에 적합하도록 객체를 생성하는 기능

```python
frame = pd.DataFrame(np.arange(9).reshape((3,3)),
                     index = ["a", "c", "d"],
                     columns = ["Ohio", "Texas", "California"])
frame
```

```python
# Reindexing Rows (default)
frame2 = frame.reindex(index = ["a", "c", "d", "d"])
frame2
```

```python
# Reindexing Columns
states = ["Texas", "Utah", "California"]
frame.reindex(columns = states)
# or equivalently
frame.reindex(states, axis = "columns")
```

### 2. 행과 열의 삭제 (Deletion)

- drop 함수

```python
data = pd.DataFrame(np.arange(16).reshape((4, 4)),
                    index = ["Ohio", "Colorado", "Utah", "New York"],
                    columns = ["one", "two", "three", "four"])
data
```

```python
# delete rows (equivalent code)
data.drop(index = ["Colorado", "Ohio"])
```

```python
# delete columns (equivalent codes)
data.drop(columns = ["two"])
data.drop("two", axis = 1)
data.drop("two", axis = "columns")
```

### 3. 색인 (Index)

- []를 통해 원하는 데이터를 선택할 수 있음

```python
data = pd.DataFrame(np.arange(16).reshape((4, 4)),
                    index=["Ohio", "Colorado", "Utah", "New York"],
                    columns=["one", "two", "three", "four"])
data
data["two"]
data[["three", "one"]]
```

* 슬라이싱(slicing)이나 조건문(Boolean 연산자)을 통해 원하는 데이터를 선택할 수 있음

```python
data[:2] 
data[data["three"] > 5]
```

- 색인을 통해 위치 or 조건의 값을 손쉽게 바꿀 수 있음

```python
data[data < 5] = 0
```

- []색인은 정수로 색인되어 있을 경우, 이것을 (정수색인이 있는 경우) 레이블로 인식할 수도 있고, 위치값으로 인식할 수도 있어, 혼선을 일으킬 수 있음
- .loc (레이블기준)과 iloc(위치기준) 을 사용하는 것이 보다 선호됨

```python
# .loc 레이블 기반 색인
data.loc["Colorado"]
data.loc[["Colorado", "New York"]]
data.loc["Colorado", ["two", "three"]]
data.loc[:"Utah", "two"]
```

```python
# .iloc: 위치 기반 색인
data.iloc[2]
data.iloc[[2, 1]]
data.iloc[2, [3, 0, 1]]
data.iloc[[1, 2], [3, 0, 1]]
data.iloc[:, :3][data.three > 5]
```

### 4. 산술연산 (Arithmetic)

- 서로다른 색인을 가지고 있는 객체간의 산술연산도 처리 가능

```python
df1 = pd.DataFrame(np.arange(9.).reshape((3, 3)), columns=list("bcd"),
                   index=["Ohio", "Texas", "Colorado"])
df2 = pd.DataFrame(np.arange(12.).reshape((4, 3)), columns=list("bde"),
                   index=["Utah", "Ohio", "Texas", "Oregon"])
df1
df2
```

```python
df1 + df2
df1.add(df2)
```

### 5. 함수 적용 (Apply)

- 데이터프레임의 각 행이나 열의 1차원 배열에 함수를 적용

```python
frame = pd.DataFrame(np.random.standard_normal((4, 3)),
                     columns=list("bde"),
                     index=["Utah", "Ohio", "Texas", "Oregon"])
frame

# 새로운 함수 정의 (range)
def f1(x):
    return x.max() - x.min()

frame.apply(f1) # along the row
frame.apply(f1, axis = "columns") # along the column
```

- map: 각원소에 적용할 수 있음

```python
# 숫자 (floating number)를 문자열로 변환
def my_format(x):
    return f"{x:.2f}"

frame.map(my_format)
```

### 6. 정렬 (Sort)

- 기본적으로 오름차순으로 정렬 (ascending 옵션을 통해 내림차순으로 변경 가능)
- sort_index: 인덱스에 따라 정렬

```python
frame = pd.DataFrame(np.arange(8).reshape((2, 4)),
                     index=["three", "one"],
                     columns=["d", "a", "b", "c"])
frame
frame.sort_index()
frame.sort_index(axis="columns")
frame.sort_index(axis="columns", ascending=False)
```

- sort_values: 값에 따라 정렬

```python
obj = pd.Series([4, 7, -3, 2])
obj.sort_values()
```

- 결측치가 있는 경우

```python
obj = pd.Series([4, np.nan, 7, np.nan, -3, 2])
obj.sort_values()
obj.sort_values(na_position="first")
```

- 데이터프레임의 경우 정렬 기준을 정해야함

```python
frame = pd.DataFrame({"b": [4, 7, -3, 2], "a": [0, 1, 0, 1]})
frame
frame.sort_values("b")
frame.sort_values(["a", "b"])
```

### 7. 순위 (Rank)

- 낮은 순으로, 동점은 평균으로 처리하는 것이 기본값

```python
obj = pd.Series([4, np.nan, 7, np.nan, -3, 2])
obj.sort_values()
obj.sort_values(na_position="first")
```

- 데이터프레임의 경우 각각 행과 열을 기준으로 모두 순위를 정할 수 있음

```python
frame = pd.DataFrame({"b": [4.3, 7, -3, 2], "a": [0, 1, 0, 1],
                      "c": [-2, 5, 8, -2.5]})
print(frame)
print(frame.rank(axis = "columns"))
```

- read_csv 함수의 주요 인수 (input argument)

| 인수 | 설명 |
| --- | --- |
| path | 파일위치를 나타내는 문자열 |

| keep_default_na | 기본 NA값 리스트를 사용할지 여부 (deault는 True) |
| encoding | 파일의 인코딩에 사용한 방법 설정 - 한글이 들어간 경우 주의 필요 |

### JSON 데이터 불러오기

- JSON (JavaScript Object Notation)도 널리 활용되는 파일의 형식

```python
obj = """
{"name": "Wes",
 "cities_lived": ["Akron", "Nashville", "New York", "San Francisco"],
 "pet": null,
 "siblings": [{"name": "Scott", "age": 34, "hobbies": ["guitars", "soccer"]},
              {"name": "Katie", "age": 42, "hobbies": ["diving", "art"]}]
}
"""
```

- json 라이브러리를 활용
- json.loads()함수를 이용하여 JSON파일을 파이썬 객체로 변환할 수 있음
