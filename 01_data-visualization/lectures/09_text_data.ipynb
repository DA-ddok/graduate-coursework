### 예제1 ) 다음 문서에서 `사업예산 금액` 을 찾아보기

```python
str = """
1. 입찰건명: 노후 스팸필터 서버 교체 사업
2. 사업예산: 35,000,000원 (부가세 포함)
3. 입찰방법: 일반경쟁입찰 (최저가 낙찰)
4. 입찰등록마감 일시: 2020. 4. 20(월) 14:00, (제출 : 서울상공회의소 총무팀(19층))
5. 입찰 일시 및 장소: 2020. 4. 23(목) 11:00, 소회의실4 (본소 지하2층)
6. 입찰참가자격: 서울상공회의소 회원업체로서 회원의 의무를 다한 업체, 사업개시 3년 이상인 업체, 사업설명회에 참석한 업체
"""
```

```python
str
```

```python
# 문항별로 분할
p = re.compile('\n')
m = p.split(str)
m
```

```python
# 항목의 제목과 내용 부분 분할
s = [re.split(":", entry, 2) for entry in m]
s
```

```python
# '사업예산' 항목을 찾아 그 내용을 확인
for entry in s:
    if re.search('사업예산', entry[0]):
        obj = entry[1]
obj
```

```python
obj1 = obj.split(" ")
obj1

obj2 = obj1[1].split('원')[0].replace(",", "")
obj2

int(obj2)
```

### 예제 2 ) 이메일 찾기

```python
text = """신승준 교수의 이메일 주소는 sjshin@korea.ac.kr이고, TA 이메일 주소는 sjshin4TA@gmail.com입니다.."""

text
```

```python
# 정규표현식으로 나타내기
pattern = r"[A-Za-z0-9._%+-]+@[A-Za-z0-9-]+\.[A-Za-z.]{2,5}" # 무슨의미지?
```

```python
regex = re.compile(pattern)
regex.findall(text)
```

```python
# 좀 더 나누어 보자
pattern2 = r"([A-Za-z0-9._%+-]+)@([A-Za-z0-9-]+\.[A-Za-z.]{2,5})"
regex2 = re.compile(pattern2)
email = regex2.findall(text)
email
```

```python
n = len(email)
mail_id = []
mail_dom = []
for i in range(n):
    mail_id.append(email[i][0])
    mail_dom.append(email[i][1])
```

```python
print(mail_id)
print(mail_dom)
```

## 분석과 무관한 단어 (Stop Words) 제거

# 4. 텍스트 자료의 표현 (Text Representation)

## Bag of Words (BoW)

## 문서단어행렬 (Document-Term Matrix; DTM)

```python
from sklearn.feature_extraction.text import CountVectorizer

sentences = ["The cat in the hat", "The dog in the house", "The bird in the sky"]
vectorizer = CountVectorizer()
bow = vectorizer.fit_transform(sentences)
```

```python
# List of Words
vectorizer.get_feature_names_out()
```

```python
# BOW Matrix
bow.toarray() # Document-Term Matrix
```

```python
# N-Gram
# Bi-gram
vectorizer2 = CountVectorizer(ngram_range=(1, 2))
bow2 = vectorizer2.fit_transform(sentences)
vectorizer2.get_feature_names_out()
bow2.toarray()
```

```python
# Tri-gram
vectorizer3 = CountVectorizer(ngram_range=(1, 3))
bow3 = vectorizer3.fit_transform(sentences)
vectorizer3.get_feature_names_out()
bow3.toarray()
```

## TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

docs = ["The cat jumped",
        "The white tiger roared",
        "Bird flying in the sky"]
docs
```

```python
vectorizer = TfidfVectorizer(norm = None)
tfidf = vectorizer.fit_transform(docs)
```

```python
vectorizer.get_feature_names_out()
```

```python
tfidf.toarray() # Document-Term Matrix
```
