## 1.1.3. Assignment operator

할당문(assignment statement)은 이름을 객체에 (재)바인딩하거나, 가변 객체의 속성이나 항목을 수정하는 데 사용된다. 간단히 말해, 단순 할당은 다음과 같이 동작한다:


1.   오른쪽에 있는 표현식이 평가되어 해당 객체가 생성되거나 획득된다.
2.   왼쪽에 있는 이름이 오른쪽의 객체에 할당되거나 바인딩된다.

하나의 객체에 여러 이름이 바인딩될 수 있다:


[1,2,3]이라는 객체에 똑같은 a라는 이름이 바인딩된 것

````python
a = [1, 2, 3]
b = a
a
````

````python
b
````

````python
a is b
````

````python
b[1] = 'hi!'
b
````

````python
a
````

````python
a = [1, 2, 3]
a
````

````python
id(a)  # id() returns the memory address of the object a
````

여기서 핵심 개념은 가변(mutable) vs. 불변(immutable)이다.


*   가변 객체는 in-place 변경이 가능하다.
*   불변 객체는 한 번 생성되면 수정할 수 없다.

## 1.2.5. Advanced iteration

모든 종류의 시퀀스(문자열, 리스트, 딕셔너리의 키 등)에 대해 반복(iterate)할 수 있다:

````python
vowels = 'aeiouy'
for i in 'powerful':
    if i in vowels:
        print(i)
````

````python
message = "Hello how are you?"
message.split()    # returns a list 
#split() 메서드는 문자열을 공백으로 나누어 리스트로 반환함
````

````python
for word in message.split():
    print(word)
````

대부분의 프로그래밍 언어들은 정수나 인덱스 이외의 것들에 대해 반복(loop)하는 것을 허용하지 않는다.   
하지만 파이썬에서는 굳이 신경 쓰지 않아도 되는 인덱스를 사용하지 않고도, 관심 있는 객체들 자체를 직접 반복할 수 있다.   
이러한 기능은 코드를 더욱 가독성 있게 만드는 데 중요한 역할을 한다.

자주 수행되는 작업 중 하나는 시퀀스의 항목 번호를 추적하면서 반복(iterate)하는 것이다:

````python
words = ('cool', 'powerful', 'readable')
for i in range(0, len(words)):
    print((i, words[i]))
````

### 특히 파이썬은 이를 위해 내장 함수인 `enumerate`를 제공한다: index와 item을 같이 리턴

````python
for index, item in enumerate(words): 
    print((index, item))
````

딕셔너리에 대하여 반복하기:

key와 value 모두로 for/while문을 반복하려면 `items()`를 사용한다.

````python
d = {'a': 1, 'b':1.2, 'c':1j}
for key, val in sorted(d.items()):
    print('Key: %s has value: %s' % (key, val))
````

````python
d = {'a': 1, 'b':1.2, 'c':1j}
d.items()
````

## 1.3.4. Pssing by object reference

변수를 함수에 전달할 때 파이썬은 해당 변수가 가리키는 객체에 대한 참조(object reference)를 전달하는 것이지,  
변수 값 자체를 전달하는 것은 아니다.

함수에 전달된 값이 불변(immutable) 객체에 대한 참조인 경우, 함수는 호출자의 변수를 수정할 수 없다.  
반면, 값이 가변(mutable) 객체에 대한 참조인 경우에는 함수가 호출자의 변수에 대한 in-place 수정이 가능하다.

````python
def try_to_modify(x, y, z):
    x = 23
    y.append(42)
    z = [99]    # new reference  : 새롭게 만들어주면 기존의 참조는 끊어지고 새로운 참조가 만들어짐
    print(x)
    print(y)
    print(z)

a = 77      # immutable variable
b = [99]    # mutable variable
c = [28]
try_to_modify(a, b, c)
````

````python
print(a)
print(b) # 가변 객체는 함수 연산 후 값이 변해있음 
print(c)
````

함수는 로컬 네임스페이스(local namespace)라 불리는 지역 변수(local variable)의 테이블을 가진다. 위의 예에서 변수 `x`는 오직 `try_to_modify` 함수 내에서만 존재한다.

## Exercise

**Exercise 1:** Write a function that displays the `n` first terms of the **Fibonacci sequence**, defined by:
$$
U_0 = 0, U_1 = 1, U_{n+2} = U_{n+1} + U_n
$$

````python
# 내풀이 : 근데 재귀호출로 기하급수적으로 늘어남 
def fibonacci(n):
    """ Return the nth Fibonacci number. """
    if n== 0:
        return 0 
    elif n== 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
````

````python
fibonacci(10)
````

````python
#교수님
def fibonacci_prof(n):
    a = 0; b = 1
    for i in range(n):
        print(a, end = ' ') # a 값이 출력되고 띄운다 
        a, b = b, a + b # a는 b의 값을 참조하고, b는 a와 b의 합을 참조하게 됨. 동시에 업데이트됨
fibonacci_prof(10)
````

````python
def fibonacci_2(n):
    fib = [0,1]
    for i in range(n-2): # 이미 0과 1이 있으므로 n-2번 반복
        fib.append(fib[i] + fib[i+1])
    return fib
````

````python
fibonacci_2(10)
````

# 1.4. Object-oriented programming (OOP)

파이썬은 객체 지향 프로그래밍(OOP)을 지원한다. OOP의 목적은 다음과 같다:

*   코드를 체계적으로 구성하고,
*   유사한 상황에서 코드를 재사용하기 위해서이다.

간단한 예로, 여러 개의 사용자 정의 함수(메서드, methods)와 변수(속성, attributes)를 하나로 묶어 놓은 새로운 객체인 Student 클래스(class)를 생성하고, 이를 활용해보자.

````python
class Student: 
    univ = 'Korea University'  # 클래스 속성으로 모든 Student 객체가 공유하는 univ 속성을 정의함. 
    #이 경우 모든 학생은 'Korea University'에 속하게 됨
    
    
    def __init__(self, name):  # self는 관습적인 인자(객체 그 자체), name은 객체가 생성될 때 전달되는 인자
        self.name = name
    def set_age(self, age): #age 라는 인자를 받아서 age라는 속성에 저장하는 메서드 
        self.age = age
    def set_major(self, major):
        self.major = major
````

````python
anna = Student('anna') #init method이 호출되어 name 속성이 'anna'로 설정된 Student 객체가 생성됨
````

````python
anna # student 클래스 
````

````python
anna.name
````

````python
anna.set_age(21)
anna.set_major('physics')
````

````python
print(anna.age); print(anna.major)
````

````python
anna.univ
````

속성(attributes)으로는 `name`, `age`, `major`를 가진다.  
이러한 메서드와 속성은 다음과 같은 표기법으로 호출할 수 있다: `classinstance.method` 또는 `classinstance.attribute`.

이번에는 이전 클래스와 동일한 메서드와 속성을 가지면서 추가로 internship이라는 속성을 포함하는 새로운 클래스인 MasterStudent를 만들고 싶다고 가정해보자. 이 경우 이전 클래스를 복사하지 않고도 클래스 상속(class inheritance)를 통하여 클래스를 생성할 수 있다:

만약 `MasterStudent`가 자기만의 `__init__` 메서드를 갖게 하고 싶다면 어떻게 할까? 이 경우에는 새로운 `__init__` 안에서 `super().__init__(...)`를 사용하여 부모 클래스의 생성자를 호출할 수 있다. 이렇게 하면 `Student`에서 수행하던 모든 초기화 작업, 예를 들어 `name`을 설정하는 작업은 그대로 유지하면서, `MasterStudent`에만 필요한 추가 속성을 더할 수 있다.”

## 1.5.4. '\_\_main\_\_' and module loading

파이썬 파일이 스크립트로서 직접 실행될 때는 특정 코드가 실행되기를 원하지만, 모듈로 import될 때는 실행되지 않기를 원할 때가 있다.  
이 경우  `if __name__ == '__main__'` 구문을 사용하면 해당 모듈이 직접 실행되었는지 확인할 수 있다.

`demo2.py`:

```
def print_b():
    "Prints b."
    print("b")


def print_a():
    "Prints a."
    print("a")


# print_b() runs on import
print_b()

if __name__ == "__main__":
    # print_a() is only executed when the module is run directly.
    print_a()
```

`demo2.py`를 import:

````python
import demo2
````

````python
import demo2  #이미 demo2 모듈이 import되어 있기 때문에 변경된 내용이 반영되지 않음.
````

## 1.5.5. Scripts or modules? How to organize your code

**Rule of thumb:**


*   여러 번 반복해서 호출되는 명령어 집합은 코드의 재사용성을 높이기 위해 함수로 작성해둔다.

*    여러 스크립트에서 공통적으로 호출되는 함수는 모듈로 작성하고, 각 스크립트에서는 그 모듈만 import하여 사용한다 (각 스크립트마다 동일한 함수를 복사하여 붙여넣는 것을 피한다!).

`import mymodule` 문이 실행되면, 파이썬은 `mymodule`이라는 모듈을 특정 디렉토리 목록에서 검색한다.  
이 목록에는 `/usr/lib64/python3.11`과 같은 설치 환경에 따라 결정되는 기본 경로들과, 환경 변수 `PYTHONPATH`에 의해 지정된 디렉토리들이 포함된다.

파이썬이 모듈을 검색할 때 사용하는 디렉토리 목록은 `sys.path` 변수에 저장되어 있다.

````python
import sys
sys.path
````

모듈은 반드시 검색 경로 내에 위치해야 한다. 따라서 이미 검색 경로에 정의된 디렉토리 안에 자신만의 모듈을 작성하거나, 파이썬 스크립트 내에서 `sys.path` 변수를 직접 수정한다.

```
import sys
new_path = '/home/junsoukchoi/das510'
if new_path not in sys.path:
    sys.path.append(new_path)
```

패키지는 서브모듈(submodule)을 가진 모듈이며, 이 서브모듈들도 다시 자신의 서브모듈을 가질 수 있다.  
특별한 파일인 `__init__.py` (빈 파일이어도 됨)는 해당 디렉토리가 파이썬 패키지임을 나타내며, 이를 통해 패키지 내 모듈을 불러올 수 있게 한다.

````python
import scipy as sp
sp.__file__

# __init__.py 파일은 패키지의 초기화 코드가 들어있는 파일로, 패키지가 import될 때 자동으로 실행됨.
````

## 1.5.7. Good practices

*   **의미 있는** 객체 **이름**을 사용하라.

*   **들여쓰기(indentation)는 필수!** 콜론(`:`) 뒤에 오는 모든 코드 블록은, 콜론이 있는 이전 줄보다 한 수준 더 들여쓰기 해야 한다.

*   **들여쓰기의 깊이:** 원하는 개수의 공백(1, 2, 3, 4, 등)으로 들여쓰기할 수 있지만, 일반적으로 4칸 들여쓰기를 사용하는 것이 좋은 관례로 여겨진다.

*   **긴 줄:** 한 줄이 너무 길어지지 않도록 (예를 들어 80자 이상으로) 작성하는 것이 좋다. 긴 줄은 `\` 문자로 줄을 나눌 수 있다.

````python
long_line = "Here is a very very long line \
that we break in two parts."
````

*   **적절한 띄어쓰기**는 가독성을 좋게 한다: 쉼표 뒤, 산술 연산자 주위 등에 공백을 넣어라
